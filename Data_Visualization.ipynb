5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility

# Generate random data
n_samples = 100
x = np.random.rand(n_samples)
y = 2 * x + np.random.normal(0, 0.1, n_samples)  # y is correlated with x
z = 3 * x + np.random.normal(0, 0.1, n_samples)  # z is also correlated with x

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Compute the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Adding titles
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Step 1: Generate sample sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Step 2: Pivot the DataFrame to create a matrix for the heatmap
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Step 3: Create the heatmap using Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Average Sales'})
plt.title('Sales Heatmap: Variation Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day of the Month')
plt.show()

# Step 4: Create the heatmap using Plotly
fig = px.imshow(
    heatmap_data.values,
    x=heatmap_data.columns,
    y=heatmap_data.index,
    color_continuous_scale='YlGnBu',
    labels=dict(x="Month", y="Day", color="Average Sales"),
    title='Sales Heatmap: Variation Across Months and Days'
)

# Show the Plotly heatmap
fig.show()

```

4. Using the given x and y data, generate a 3D surface plot to visualize the function

z=sin(sqrt(x^2+y^2))

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Step 1: Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Step 2: Create a DataFrame (optional, but good for consistency)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Step 3: Create the surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x.flatten(), y=y.flatten())])

# Step 4: Update layout
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(
                      xaxis_title='X axis',
                      yaxis_title='Y axis',
                      zaxis_title='Z axis'),
                  autosize=True)

# Step 5: Show the plot
fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Optional: If you are using a Jupyter Notebook
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Compute the sine of x
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label='sin(x)', line_width=2, line_color='blue')

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Optional: If you are using a Jupyter Notebook
output_notebook()

# Generate random data
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values between 0 and 100
y = np.random.rand(n) * 100  # Random y values between 0 and 100
sizes = np.random.randint(5, 50, n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a Bokeh figure
p = figure(title="Scatter Plot with Randomly Generated Data",
           x_axis_label='X Values',
           y_axis_label='Y Values',
           width=800,
           height=400)

# Add a scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', source=source, fill_alpha=0.6)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Optional: If you are using a Jupyter Notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits',
           y_axis_label='Counts',
           width=800,
           height=400)

# Add a bar renderer
p.vbar(x='fruits', top='counts', source=source, width=0.4, color="orange")

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Optional: If you are using a Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           width=800,
           height=400)

# Add a quad renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color="blue", line_color="black", alpha=0.7)

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper

# Optional: If you are using a Jupyter Notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid for the heatmap
xx, yy = np.meshgrid(x, y)

# Prepare the data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

# Create a Bokeh figure
p = figure(title="Heatmap of Random Data",
           x_axis_label='X',
           y_axis_label='Y',
           width=400,
           height=400)

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data['value'].min(), high=data['value'].max())

# Create the heatmap using squares
p.rect(x='x', y='y', width=0.1, height=0.1, source=data,
        fill_color=linear_cmap('value', palette="Viridis256", low=data['value'].min(), high=data['value'].max()),
        line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
